```python
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
```

```python
def read_columns_as_arrays(file_path):
    # Read the .txt file using pandas, with semicolon as separator
    df = pd.read_csv(file_path, sep=';', header=None)
    
    # Extract each column as a NumPy array
    col1_array = df[0].to_numpy()
    col2_array = df[1].to_numpy()
    
    return col1_array, col2_array
```

```python
steps, value = read_columns_as_arrays("/Users/pedrobraga/Documents/Cambridge/cpp/practical_three/output_euler.txt")
```

```python
print(value)
```

```python
plt.plot(steps, value)
```
